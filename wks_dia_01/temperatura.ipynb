```python
import numpy as np
import matplotlib.pyplot as plt

max = [29.3, 29.3, 29.3, 28.9, 28.2, 27.1, 26.5, 26.7, 27.5, 28.5, 29.2, 29.3]
med = [26.6, 26.7, 26.8, 26.4, 25.9, 24.9, 24.3, 24.3, 24.9, 25.8, 26.4, 26.6]
min = [24.6, 24.8, 24.8, 24.6, 24.1, 23.2, 22.6, 22.4, 23, 23.7, 24.3, 24.6]
 
barWidth = 0.30
plt.figure(figsize=(10,5))
r1 = np.arange(len(max))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
plt.bar(r1, max, color='#FF0505', width=barWidth, label='Max')
plt.bar(r2, med, color='#FF4705', width=barWidth, label='Med')
plt.bar(r3, min, color='#2879DC', width=barWidth, label='Min')

plt.xlabel('Meses')
plt.xticks([r + barWidth for r in range(len(max))], ['Jan', 'Fev', 'Mar', 'Ab', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
plt.ylabel('Temperatura')
plt.title('Gráfico de temperatura por mês em João Pessoa')
plt.legend()
plt.show()
```
